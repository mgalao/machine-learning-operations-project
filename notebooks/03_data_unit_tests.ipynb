```python
df_ingested = pd.read_csv('../data/02_intermediate/ingested_data.csv')
df_ingested.info()
```

```python
sns.set(style="whitegrid")

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

# Age distribution
sns.histplot(df_ingested["age"], bins=50, kde=True, ax=axs[0], color='skyblue')
axs[0].set_title("Age Distribution")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Frequency")

# cc_num length distribution
cc_num_lengths = df_ingested["cc_num"].astype(str).str.len()
cc_num_len_counts = cc_num_lengths.value_counts().sort_index()
axs[1].bar(cc_num_len_counts.index.astype(str), cc_num_len_counts.values, color='mediumseagreen')
axs[1].set_title("cc_num Length Distribution")
axs[1].set_xlabel("Length of cc_num")
axs[1].set_ylabel("Count")

for i, v in enumerate(cc_num_len_counts.values):
    axs[1].text(i, v + max(cc_num_len_counts.values)*0.01, str(v), ha='center', va='bottom')

# ZIP code length distribution
zip_lengths = df_ingested["zip"].astype(str).str.len()
zip_len_counts = zip_lengths.value_counts().sort_index()
axs[2].bar(zip_len_counts.index.astype(str), zip_len_counts.values, color='coral')
axs[2].set_title("ZIP Code Length Distribution")
axs[2].set_xlabel("Length of ZIP Code")
axs[2].set_ylabel("Count")

for i, v in enumerate(zip_len_counts.values):
    axs[2].text(i, v + max(zip_len_counts.values)*0.01, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## age

```python
# Find rows where age is outside expected range
df_ingested.loc[(df_ingested["age"] < 16) | (df_ingested["age"] > 120), ["age"]]
```

## cc_num

```python
# 13 to 19 digit number
regex = r"^\d{13,19}$"

# Check cc_num that do NOT match regex
df_ingested.loc[~df_ingested["cc_num"].astype(str).str.match(regex), ["cc_num"]]
```

## zip

```python
# 5 or 5+4 digit ZIP code pattern
zip_regex = r"^\d{5}(-\d{4})?$"

# Check invalid ZIPs
df_ingested.loc[~df_ingested["zip"].astype(str).str.match(zip_regex), ["zip"]]
```

## merch_zipcode

```python
# Show nulls in merch_zipcode
df_ingested.loc[df_ingested["merch_zipcode"].isnull(), ["merch_zipcode"]]

# Percentage null
null_rate = df_ingested["merch_zipcode"].isnull().mean()
print(f"Null % in 'merch_zipcode': {null_rate:.2%}")
```
